```python
#Importing classes
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
#Loading Data
df_Alim = pd.read_excel("C:/Users/Alim/OneDrive/Desktop/COMP341/Lab1/Real estate valuation data set.xlsx")

#Data Info
df_Alim.head()
df_Alim.columns.values
df_Alim.isnull().sum()
df_Alim.describe()
df_Alim.columns
df_Alim.info()

#Split the real estate data into training and testing datasets. Targeting "Y house price of unt area" within dataset
X_train, X_test, y_train, y_test = train_test_split(df_Alim.drop("Y house price of unit area", axis=1), df_Alim["Y house price of unit area"], test_size=0.2)

#Training the linear regression model
linearReg = LinearRegression().fit(X_train, y_train)

#Prediction for the data
y_prediction = linearReg.predict(X_test)

#Finding coefficient of the linear Regression model
r2 = r2_score(y_test, y_prediction)

#Printing the coefficient of the model
print(f"coefficient of determination (R2): {r2:.2f}")

#Plot the results
plt.scatter(y_test, y_prediction)
plt.title("Linear Regression")
plt.xlabel("Values")
plt.ylabel("Prediction")
plt.show()
```
